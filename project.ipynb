```python
import pandas as pd
```

```python
train1 = pd.read_csv('Train_Data.csv')
train1.head()
```

```python
new_train1 = train1.drop(['revenue','date','campaign'],axis=1)
new_train1
```

```python
new_train1.shape
```

```python
test1 = pd.read_csv('Test_Data.csv')
test1.head()
```

```python
new_test1 = test1.drop(['date','campaign'],axis=1)
new_test1
```

```python
new_test1.shape
```

```python
new_test_cat = pd.get_dummies(new_test1)
new_test_cat.head()
```

```python
col = new_test_cat.columns
col
```

```python
new_train_cat = pd.get_dummies(new_train1)
new_train_cat
```

```python
col_train = new_train_cat.columns
col_train
```

```python
new_train_cat[col]
```

```python
new_test_cat.drop(['ad_ad 73', 'ad_ad 71', 'ad_ad 72', 'ad_ad 75', 'ad_ad 74'],axis=1,inplace = True)
```

```python
new_test = new_test_cat
new_test
```

```python
col_test = new_test_cat.columns
col_test
```

```python
new_train = new_train_cat[col_test]
new_train
```

```python
X = new_train
X
```

```python
y = train1.revenue
y
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor(n_estimators= 1000, learning_rate = 0.01)
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
pred = model.predict(new_test)
```

```python
pred = pd.DataFrame(pred)
pred
```

```python
pred[pred<15] = 0
pred
```

```python
columns={0:'revenue'}
pred.rename(columns=columns, inplace=True)
pred
```

```python
pred.to_csv('rev_final.csv', index=False)
```
